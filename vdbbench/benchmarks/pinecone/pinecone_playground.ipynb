```python
glove100_dataset = load_dataset("ANN_GloVe_d100_angular")
glove100_dataset.head()
```

```python
import pandas as pd
# glove100_dataset["values"]
```

```python
pc = Pinecone(api_key=PINECONE_API_KEY)
pc.create_index(
    name="glove100d",
    dimension=100,
    metric="cosine",
    # change later, serverless only available on AWS 
    # trying to start with some boilerplate code rn git commit -am ""
    spec=ServerlessSpec(
        cloud='aws', 
        region='us-west-2'
    ) 
) 

index = pc.Index("glove100d")

index.upsert_from_dataframe(glove100_dataset.drop(columns=["sparse_values","metadata","blob"]))
```
